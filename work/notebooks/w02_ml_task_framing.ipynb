## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Predicting house values cannot be done accurately using simple if-else rules because house prices depend on many factors working together. For example, median income, location, population, total rooms, and housing age all influence the final house value.

A fixed rule might only consider one or two factors, which would lead to inaccurate predictions. Machine Learning can learn complex relationships between multiple features from historical data and use those patterns to make better predictions for new housing records.

Therefore, Machine Learning provides more flexible and accurate decision support than a fixed rule-based approach.
