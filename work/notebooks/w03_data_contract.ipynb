### Label

### Context
1. report_date
2. client_hash_id
3. content_hash_id

These fields identify the content and the reporting date but are not directly used as predictive features.

### Excluded
1. client_has_gsc
2. client_has_ga4

These fields indicate whether a client has connected Google Search Console or Google Analytics. They are excluded because they describe system setup rather than content performance and may introduce unnecessary bias.
